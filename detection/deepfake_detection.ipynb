```python
# 1. Setup
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
import os

# 2. Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 3. Paths
from google.colab import drive
drive.mount('/content/drive')

data_dir = '/content/drive/MyDrive/Colab Notebooks/deepfake_detection/dataset/train'
save_dir = '/content/drive/MyDrive/Colab Notebooks/deepfake_detection/models'
os.makedirs(save_dir, exist_ok=True)

# 4. Data Transforms
data_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# 5. Dataset and Dataloader (Only Train)
train_dataset = datasets.ImageFolder(data_dir, transform=data_transforms)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)

# 6. Load Pretrained ResNet18
model = models.resnet18(pretrained=True)

# Replace the final layer for 2 classes (real vs fake)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)
model = model.to(device)

# 7. Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 8. Training loop
num_epochs = 10  # You can change this

losses = []

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for inputs, labels in train_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    losses.append(epoch_loss)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}")

# 9. Save the model
torch.save(model.state_dict(), os.path.join(save_dir, 'resnet18_deepfake_v3.pth'))
print(f"Model saved to {save_dir}")

# 10. Save loss values to CSV (for graphs later)
import pandas as pd
loss_df = pd.DataFrame(losses, columns=["Loss"])
loss_df.to_csv(os.path.join(save_dir, "training_losses.csv"), index=False)
print(f"Losses saved to {save_dir}")
```

```python
# 1. Reload model
model = models.resnet18(pretrained=False)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)
model.load_state_dict(torch.load(os.path.join(save_dir, 'resnet18_deepfake.pth')))
model = model.to(device)

# 2. Reload optimizer (optional but better for smooth continuation)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 3. Load previous loss values
loss_df = pd.read_csv(os.path.join(save_dir, "training_losses.csv"))
losses = list(loss_df["Loss"])

# 4. Resume training
additional_epochs = 10
start_epoch = len(losses)  # continue from previous

for epoch in range(start_epoch, start_epoch + additional_epochs):
    model.train()
    running_loss = 0.0

    for inputs, labels in train_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    losses.append(epoch_loss)
    print(f"Epoch [{epoch+1}/{start_epoch+additional_epochs}], Loss: {epoch_loss:.4f}")

# 5. Save updated model and loss
torch.save(model.state_dict(), os.path.join(save_dir, 'resnet18_deepfake_v2.pth'))
print(f"Updated model saved to {save_dir}")

# 6. Save updated losses to CSV
loss_df = pd.DataFrame(losses, columns=["Loss"])
loss_df.to_csv(os.path.join(save_dir, "training_losses.csv"), index=False)
print(f"Updated losses saved to {save_dir}")
```

```python
# THIS
# IMAGE INFERENCING
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import os

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load your model
model = models.resnet18(weights=None)  # No pre-trained weights
model.fc = nn.Linear(model.fc.in_features, 2)  # 2 classes: real and fake
model.load_state_dict(torch.load('/content/drive/MyDrive/Colab Notebooks/deepfake_detection/models/resnet18_deepfake_v3.pth', map_location=device))
model = model.to(device)
model.eval()

# Define same transforms you used during training
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],  # Same ImageNet normalization
                         [0.229, 0.224, 0.225])
])

# Prediction function
def predict_image(image_path, threshold=0.9):
    img = Image.open(image_path).convert('RGB')
    img = transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(img)
        probs = torch.softmax(outputs, dim=1)

    fake_prob = probs[0][0].item()
    real_prob = probs[0][1].item()

    print(f"Prediction probabilities: Fake={fake_prob:.4f}, Real={real_prob:.4f}")

    if fake_prob >= real_prob:
        return "Fake"
    else:
        return "Real"



# Example: Predict on an image
image_path = '/content/drive/MyDrive/Colab Notebooks/deepfake_detection/testImages/test_img_9.jpg'
prediction = predict_image(image_path)
print(f'The model predicts that the image is: {prediction.upper()}')
```

```python
# THIS
# VIDEO INFERENCING
import cv2
import torch
from torchvision import transforms

# Function to predict on every nth frame of a video and make final decision
def predict_video_final(video_path, n=5):
    # Open the video
    cap = cv2.VideoCapture(video_path)

    # Check if the video was opened successfully
    if not cap.isOpened():
        print(f"Error opening video file {video_path}")
        return

    # Get frame rate and frame count
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Video has {frame_count} frames at {fps} fps.")

    frame_idx = 0
    predictions = {'real': 0, 'fake': 0}  # Dictionary to store count of real and fake predictions

    while True:
        ret, frame = cap.read()
        if not ret:
            break  # Exit loop when all frames are processed

        if frame_idx % n == 0:  # Process every nth frame (e.g., every 5th frame)
            frame_resized = cv2.resize(frame, (224, 224))  # Resize to match model input size
            frame_rgb = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2RGB)  # Convert to RGB
            frame_tensor = transforms.ToTensor()(frame_rgb).unsqueeze(0).to(device)  # Convert to tensor and send to device

            # Make prediction for this frame
            prediction = predict_image_tensor(frame_tensor)

            # Update the count of real and fake predictions
            if prediction == 'real':
                predictions['real'] += 1
            else:
                predictions['fake'] += 1

            # Print prediction for every nth frame
            print(f"Frame {frame_idx + 1}/{frame_count} prediction: {prediction.upper()}")

        frame_idx += 1

    # Release the video capture object
    cap.release()

    # Final decision based on majority predictions
    if predictions['fake'] > predictions['real']:
        final_prediction = "DEEPFAKE"
    else:
        final_prediction = "REAL"

    print("\n")
    print(f"Final prediction for the video: {final_prediction}")
    return final_prediction

# Helper function to predict on tensor (used for frames)
def predict_image_tensor(image_tensor):
    # Make predictions
    model.eval()
    with torch.no_grad():
        outputs = model(image_tensor)
        _, predicted = torch.max(outputs, 1)

    # Get class names (Real vs Fake)
    class_names = ['real', 'fake']

    # Get the predicted class
    predicted_class = class_names[predicted.item()]
    return predicted_class

# Example usage: Predict on every 5th frame of a single video
video_path = '/content/drive/MyDrive/Colab Notebooks/deepfake_detection/testVideos/test_video_1.mp4'
final_prediction = predict_video_final(video_path, n=5)

# The final result will be printed, and the decision is stored in final_prediction
```
